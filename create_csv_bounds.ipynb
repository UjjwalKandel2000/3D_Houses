```python
import rasterio
import time
import json
import pandas as pd
#https://downloadagiv.blob.core.windows.net/dhm-vlaanderen-ii-dtm-raster-1m/DHMVIIDTMRAS1m_k01.zip
list_url = []
for i in range(40, 44):
    if i<10:
        page_num = str(i) + ".zip"
        page_num2 = str(i)+ ".tif"
        url = (
            "https://downloadagiv.blob.core.windows.net/dhm-vlaanderen-ii-dsm-raster-1m/DHMVIIDSMRAS1m_k0" + page_num \
           +"!/GeoTIFF/DHMVIIDSMRAS1m_k0"+page_num2 )
        list_url.append(url)
    else:
        page_num = str(i) + ".zip"
        page_num2 = str(i)+ ".tif"
        url = (
                "https://downloadagiv.blob.core.windows.net/dhm-vlaanderen-ii-dsm-raster-1m/DHMVIIDSMRAS1m_k" + page_num \
            +"!/GeoTIFF/DHMVIIDSMRAS1m_k"+page_num2
        )
        list_url.append(url)
```

```python
list_url
```

```python
counter = 40
for url in list_url:
    print(counter)
    img_DTM = rasterio.open(f'zip+{url}')
    list_of_bounds.append(
{  "file_name": f'zip+{url}',
  "left": img_DTM.bounds.left,
  "bottom": img_DTM.bounds.bottom,
  "right": img_DTM.bounds.right,
  "top": img_DTM.bounds.top})
    counter+=1
    
```

```python
df = pd.DataFrame(list_of_bounds)
df
```

```python
df.to_csv("bounding_box.csv", mode="w+", index=True)
```
